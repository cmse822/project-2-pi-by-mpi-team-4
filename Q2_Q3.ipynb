```python
import matplotlib.pyplot as plt
import seaborn as sns 
import subprocess
import pandas as pd
import numpy as np
```

```python
data_q2 = pd.read_csv('Q2.csv')
data_q2.columns = ["Darts", 'processors','pi','time','error']
data_q2.head()

n_array = np.array([1,2,3,4])
```

```python
data_1p = data_q2[data_q2['processors'] == 1]
data_2p = data_q2[data_q2['processors'] == 2]
data_3p = data_q2[data_q2['processors'] == 3]
data_4p = data_q2[data_q2['processors'] == 4]
```

```python
rt_array= [np.mean(data_1p['time']),np.mean(data_2p['time']),np.mean(data_3p['time']),np.mean(data_4p['time'])]
error_q2 = [np.std(data_1p['time']),np.std(data_2p['time']),np.std(data_3p['time']),np.std(data_4p['time'])]
```

```python
ideal_array = []
for i in n_array:
    ideal_array.append(rt_array[0]/i)
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size
plt.errorbar(n_array, rt_array, yerr=error_q2, fmt='--o', color='r', label='Actual', linewidth=1.5, markersize=8)  # Increase linewidth and markersize
plt.xlabel("Number of Processors", fontsize=14)  # Increase font size
plt.ylabel("Runtime (s)", fontsize=14)  # Increase font size
plt.legend(fontsize=12)  # Increase font size for legend
plt.xticks(np.arange(1, 5, 1))
plt.title("Pt 3 Q2 with 1E6 Darts")
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines with dashed style and transparency
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.savefig("part4_p2.png", dpi=300)  # Increase resolution for better quality
plt.show() 
```
